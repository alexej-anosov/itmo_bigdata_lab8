```python
import pandas as pd

    
df = pd.read_csv('../data/en.openfoodfacts.org.products.csv', sep='\t', nrows=200_000)
df
```

```python
df = df[['code', 'product_name', *df.columns[[df[column].dtype!='object' for column in df.columns]]]]
```

```python
df = df[df.columns[df.isna().sum()<100000]]
```

```python
df = df.dropna(axis=0).reset_index(drop=True)
```

```python
df = df.sample(10000).reset_index(drop=True)
```

```python
df['created_t'] = pd.to_datetime(df['created_t'],unit='s')
df['last_modified_t'] = pd.to_datetime(df['last_modified_t'],unit='s')
df['last_updated_t'] = pd.to_datetime(df['last_updated_t'],unit='s')
df.to_csv('../data/subset.csv', index=False)
```
